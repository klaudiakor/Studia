## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
selected_homes = data.loc[(data["bedrooms"]==3) & (data["bathrooms"] == 2) & (data["floors"] == 2)]
selected_homes.describe()
#Hint: pandas.DataFrame.loc
```

There are data about houses with 3 bedrooms, 3 bathrooms and 2 floors. \\
From this table, we can tell that this kind of home costs between 200'000 and 1'700'000. Square footage is between 1020 and 3650. So the difference is pretty big despite that these houses should be similar because we had a specified number of floors and rooms. So in these bigger houses, we have probably bigger rooms and maybe more space for eg. a living room or kitchen.

```python
#b)
#Your code:
plt.figure(figsize=(12,9))
plt.scatter(data["grade"], data["price"], color = "black",s=15)
plt.scatter(selected_homes["grade"],selected_homes["price"], color = "yellow", edgecolors='k', s=70)
plt.xlabel("grade")
plt.ylabel("price")
```

The black plot is created from all data, the yellow one is from selected data (homes with 3 bedrooms, 2 bathrooms and 2 floors)
We can see that the new data are concentrated in the centre, they have grades from 6 to 10, so these are pretty nice houses.


```python
#Find such apartments with best price to footage ratio.
#find price for 1 sqft (pice/sqft)

def Calculate1SqftPrice(price, sqft):
  return price/sqft;


sqftPriceRatio = []
for i in range(0,len(selected_homes.index)):
  ratio = Calculate1SqftPrice(selected_homes["price"].values[i], selected_homes["sqft_living"].values[i])
  sqftPriceRatio.append(ratio)
  
selected_homes['sqft_price_ratio'] = sqftPriceRatio
sorted = selected_homes.sort_values(["sqft_price_ratio"], ascending = True)

sorted.head()
#Hint: .loc, .sort_values(), .index
```

There are apartments with best price to footage ratio.
The best one cost 260'000 and have 2710 square footage.
